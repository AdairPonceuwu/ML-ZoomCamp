```python
data = data.assign(above_average=df['price'])
data
```

```python
numerical = ['year', 'engine_hp', 'engine_cylinders','highway_mpg','city_mpg','above_average']
df_full_train, df_test = train_test_split(data[numerical], test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=42)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train_orig = df_train.above_average.values
y_val_orig = df_val.above_average.values
y_test_orig = df_test.above_average.values

y_train = np.log1p(y_train_orig)
y_val = np.log1p(y_val_orig)
y_test = np.log1p(y_test_orig)

del df_train['above_average']
del df_val['above_average']
del df_test['above_average']
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from math import sqrt
```

```python
alphas = [0, 0.01, 0.1, 1, 10]
rmse_scores = []
for alpha in alphas:

    ridge_model = Ridge(solver='sag', alpha=alpha, random_state=42)

    ridge_model.fit(df_train.values, y_train)

    y_pred = ridge_model.predict(df_val.values)

    rmse = sqrt(mean_squared_error(y_test, y_pred))
    
    rmse = round(rmse, 3)

    rmse_scores.append(rmse)

rmse_scores
```
