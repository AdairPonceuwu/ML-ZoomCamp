```python
import pandas as pd
import numpy as np
import wget
```

```python
url = "https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/master/chapter-02-car-price/data.csv"
df = pd.read_csv(url)
```

```python
df
```

```python
base = ['Make',
'Model',
'Year',
'Engine HP',
'Engine Cylinders',
'Transmission Type',
'Vehicle Style',
'highway MPG',
'city mpg',
'MSRP'] 


data = df[base]
```

```python
data
```

```python
data.columns = data.columns.str.replace(' ', '_').str.lower()
```

```python
data = data.fillna(0)
```

```python
data_class = data.copy()
mean = data_class['msrp'].mean()

data_class['above_average'] = np.where(data_class['msrp']>=mean,1,0)
```

```python
data_class
```

```python
data_class.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
seed=1
```

```python
df_full_train, df_test = train_test_split(data_class, test_size=0.2, random_state=seed)
df_train, df_val = train_test_split(df_full_train, test_size=0.2, random_state=seed)
```

```python
numerical = ['year', 'engine_hp', 'engine_cylinders','highway_mpg','city_mpg']
```

```python
categorical = ['make', 'model', 'transmission_type','vehicle_style']
```

```python
from sklearn.metrics import roc_auc_score
```

```python
for i in numerical:
    score = roc_auc_score(df_train.above_average, df_train[i])
    if score<0.5:
        score = roc_auc_score(df_train.above_average, -df_train[i])
        print(f"{i}: {score}")
    else:
        print(f"{i}: {score}")

```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
y_train = df_train.above_average.values
y_val = df_val.above_average.values
y_test = df_test.above_average.values

del df_train['above_average']
del df_val['above_average']
del df_test['above_average']

del df_train['msrp']
del df_val['msrp']
del df_test['msrp']
```

```python
dv = DictVectorizer(sparse=False)
train_dict = df_train[categorical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dict)

val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
decision = (y_pred >= 0.5)
(y_val == decision).mean()
```

```python
round(roc_auc_score(y_val, y_pred),3)
```

```python
scores = []

thresholds = np.arange(0.0, 1.01, 0.01)

for t in thresholds:
    actual_positive = (y_val == 1)
    actual_negative = (y_val == 0)
    
    predict_positive = (y_pred >= t)
    predict_negative = (y_pred < t)

    tp = (predict_positive & actual_positive).sum()
    tn = (predict_negative & actual_negative).sum()

    fp = (predict_positive & actual_negative).sum()
    fn = (predict_negative & actual_positive).sum()

    p = tp / (tp + fp)
    r = tp / (tp + fn)

    f_1=(2*(p*r)/(p+r))

    scores.append((t, tp, fp, fn, tn,r,p,f_1))
```

```python
columns = ['threshold', 'tp', 'fp', 'fn', 'tn', 'r','p','f_1']
df_scores = pd.DataFrame(scores, columns=columns)

df_scores['tpr'] = df_scores.tp / (df_scores.tp + df_scores.fn)
df_scores['fpr'] = df_scores.fp / (df_scores.fp + df_scores.tn)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df_scores.threshold, df_scores['r'], label='Precision')
plt.plot(df_scores.threshold, df_scores['p'], label='Recal')
plt.legend()
```

```python
from sklearn.model_selection import KFold
from tqdm.auto import tqdm
```

```python
def train(df_train, y_train, C=1.0):
    dicts = df_train[categorical + numerical].to_dict(orient='records')

    dv = DictVectorizer(sparse=False)
    X_train = dv.fit_transform(dicts)

    model = LogisticRegression(solver='liblinear', C=C, max_iter=1000)
    model.fit(X_train, y_train)
    
    return dv, model
```

```python
def predict(df, dv, model):
    dicts = df[categorical + numerical].to_dict(orient='records')

    X = dv.transform(dicts)
    y_pred = model.predict_proba(X)[:, 1]

    return y_pred
```

```python
df_full_train, df_test = train_test_split(data_class, test_size=0.2, random_state=seed)
df_train, df_val = train_test_split(df_full_train, test_size=0.2, random_state=seed)
del df_full_train['msrp']
```

```python
for C in tqdm([1]):
    kfold = KFold(n_splits=5, shuffle=True, random_state=1)

    scores = []

    for train_idx, val_idx in kfold.split(df_full_train):
        df_train = df_full_train.iloc[train_idx]
        df_val = df_full_train.iloc[val_idx]

        y_train = df_train.above_average.values
        y_val = df_val.above_average.values

        dv, model = train(df_train, y_train, C=C)
        y_pred = predict(df_val, dv, model)

        auc = roc_auc_score(y_val, y_pred)
        scores.append(auc)

    print('C=%s %.3f +- %.3f' % (C, np.mean(scores), np.std(scores)))
```

```python
for C in tqdm([0.01, 0.1, 0.5, 10]):
    kfold = KFold(n_splits=5, shuffle=True, random_state=1)

    scores_F = []

    for train_idx, val_idx in kfold.split(df_full_train):
        df_train = df_full_train.iloc[train_idx]
        df_val = df_full_train.iloc[val_idx]

        y_train = df_train.above_average.values
        y_val = df_val.above_average.values

        dv, model = train(df_train, y_train, C=C)
        y_pred = predict(df_val, dv, model)

        auc = roc_auc_score(y_val, y_pred)
        scores_F.append(auc)

    print('C=%s %.3f +- %.3f' % (C, np.mean(scores_F), np.std(scores_F)))
```
